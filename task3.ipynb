## Removing Outliers Using IQR for `age`, `campaign`, and `duration`


```python
column=df[['age','campaign','duration']]
q1=np.percentile(column,25)
q3=np.percentile(column,75)
iqr=q3-q1
lower_bound=q1-1.5*iqr
upper_bound=q3+1.5*iqr
df[['age','campaign','duration']]=column[(column>lower_bound)&(column<upper_bound)]
```

## Defining Highly Correlated Columns


```python

high_corr_cols = ['emp.var.rate', 'euribor3m', 'nr.employed']


```

## Creating a Copy of `df` and Listing Its Columns


```python
df1=df.copy()
df1.columns
```

```python

existing_cols_to_drop = [col for col in high_corr_cols if col in df1.columns]
df1.drop(columns=existing_cols_to_drop, inplace=True)


```

```python
df1.shape
```

## Label Encoding of Categorical Features


```python
from sklearn.preprocessing import LabelEncoder
lb = LabelEncoder()
df_encoded = df1.apply(lb.fit_transform)
df_encoded
```

## Checking the Value Counts of the `deposit` Column in the Encoded DataFrame


```python
df_encoded['deposit'].value_counts()
```

## Splitting the Encoded DataFrame into Features (`x`) and Target (`y`)


```python
x=df_encoded.drop('deposit',axis=1)
y=df_encoded['deposit']
print(x.shape)
print(y.shape)
print(type(x))
print(type(y))
```

## Splitting Data into Training and Test Sets for Model Training


```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test,=train_test_split(x,y,test_size=0.25,random_state=1)
print(x_train.shape)
print(x_test.shape)
print(y_train.shape)
print(x_test.shape)
      
```

## Defining Model Evaluation Functions: Accuracy, Confusion Matrix, and Scores


```python
from sklearn.metrics import confusion_matrix,classification_report,accuracy_score
def eval_model(y_test,y_pred):
    acc=accuracy_score(y_test,y_pred)
    print('Accuracy_Score',acc)
    cm=confusion_matrix(y_test,y_pred)
    print('Confusion Matrix\n',cm)
    print('Classification Report\n',classification_report(y_test,y_pred))

def mscore(model):
    train_score=model.score(x_train,y_train)
    test_score=model.score(x_test,y_test)
    print('Training Score',train_score)
    print('Testing Score',test_score)
    
```

## Training a Decision Tree Classifier 


```python
from sklearn.tree import DecisionTreeClassifier
dt=DecisionTreeClassifier(criterion='gini',max_depth=5,min_samples_split=10)
dt.fit(x_train,y_train)
```

## Evaluating the Performance of the Trained Decision Tree Classifier


```python
mscore(dt)
```

## Making Predictions with the Trained Decision Tree Classifier


```python
ypred_dt=dt.predict(x_test)
print(ypred_dt)
```

## Evaluating the Predictions of the Decision Tree Classifier


```python
eval_model(y_test,ypred_dt)
```

## Preparing for Decision Tree Visualization: Feature Names and Class Labels


```python
from sklearn.tree import plot_tree
cn=['no','yes']
fn=x_train.columns
print(fn)
print(cn)
```

## Visualizing the Trained Decision Tree Classifier


```python
plt.figure(figsize=(30,10))
plot_tree(dt,class_names=cn,filled=True)
plt.show()
```
